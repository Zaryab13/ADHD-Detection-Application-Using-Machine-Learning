```python
import os
import scipy.io as sio
import pandas as pd

# Replace 'input_folder_path' with the path to your folder containing .mat files
input_folder_path = 'datasets/ADHD Data Set (EEG)/ADHD_part1'

# Replace 'output_folder_path' with the path to your desired output folder for CSV files
output_folder_path = 'datasets/ADHD Data Set (EEG)/ADHD_part1CSV'

# Iterate through all files in the input folder
for file_name in os.listdir(input_folder_path):
    if file_name.endswith('.mat'):
        input_file_path = os.path.join(input_folder_path, file_name)
        
        # Load the .mat file
        data = sio.loadmat(input_file_path)
        
        # Get the variable key (which is the filename without the extension)
        variable_key = os.path.splitext(file_name)[0]
        
        # Access the variable data and convert it to a DataFrame
        df = pd.DataFrame(data[variable_key])
        
        # Specify the output CSV file path in the output folder
        output_file_path = os.path.join(output_folder_path, variable_key + '.csv')
        
        # Save DataFrame to a CSV file in the output folder
        df.to_csv(output_file_path, index=False)
```

```python
import os

# Replace 'folder_path' with the path to your folder containing .csv files
folder_path = 'datasets/ADHD Data Set (EEG)/ADHD_part1'

# Get a list of all files in the folder
file_list = os.listdir(folder_path)

# Filter out only the .csv files
csv_files = [file for file in file_list if file.endswith('.csv')]

# Iterate through the list of .csv files and delete them
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    os.remove(file_path)
    print(f"Deleted: {file}")
```
